### Data Preprocessing: 

```python
data = data.replace('?', np.nan)
data['horsepower'] = data['horsepower'].astype(float)
```

```python
data = data.dropna()
```

Data preprocessing ensures the dataset is clean and suitable for modeling. Steps include replacing missing values (handled by `data.replace('?', np.nan)` and converting `'horsepower'` to a numeric type (`data['horsepower'] = data['horsepower'].astype(float)`). Dropping rows with missing values (`data = data.dropna()`) ensures data integrity for reliable model training.

### Define Target Variable (y) and Feature Variables (x):

```python
data.columns
```

```python
y = data['mpg']
y.shape
x = data[['displacement' ,'horsepower', 'weight' , 'acceleration']]
x.shape
```

```python
x
```

The target variable (`y`) is defined as `data['mpg']`, representing the MPG values to be predicted. Feature variables (`x`) include `'displacement'`, `'horsepower'`, `'weight'`, and `'acceleration'`, selected based on their potential influence on MPG.

### Train Test Split:

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test, y_train, y_test = train_test_split(x, y, train_size =0.7 , random_state=2529) 
```

```python
x_train.shape , x_test.shape, y_train.shape, y_test.shape
```

The dataset is split into training and testing sets using `train_test_split()` from `sklearn.model_selection.` This division (`train_size=0.7`) allocates 70% of the data for training the model and 30% for testing its performance. A `random_state` parameter ensures reproducibility of results.

### Modeling:

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

A linear regression model (`LinearRegression` from `sklearn.linear_model`) is chosen for its simplicity and interpretability. The model is trained (`model.fit(x_train, y_train)`) on the training data to establish a mathematical relationship between feature variables (`x_train`) and the target variable (`y_train`).

### Model Evaluation:

```python
from sklearn.metrics import mean_squared_error
```

```python
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'Root Mean Squared Error: {rmse}')
```

The trained model (`model`) predicts MPG values (`y_pred`) on the test set (`x_test`). The mean squared error (MSE) is computed using `mean_squared_error(y_test, y_pred)` to quantify the model's predictive performance. The root mean squared error (RMSE) provides a more interpretable measure of prediction accuracy.

### Prediction:

```python
y_pred = model.predict(x_test)
```

Finally, the model generates predictions (`y_pred`) for MPG values on the test data (`x_test`). These predictions are essential for assessing how well the model generalizes to unseen data, indicating its practical utility for predicting vehicle fuel efficiency.
